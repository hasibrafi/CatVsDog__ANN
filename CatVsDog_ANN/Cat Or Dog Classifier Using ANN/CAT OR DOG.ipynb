```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
dogs = os.listdir('./Dogs')
cats = os.listdir('./Cats')
```

```python
dogs[:10]
```

```python
x = []
y = []
```

```python
for i in dogs:
    img = cv2.imread('./dogs/'+i)
    img = cv2.resize(img, (28, 28))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = img.flatten()
    x.append(img)
    y.append(1)
    #print(img.shape)
```

```python
for i in cats:
    img = cv2.imread('./cats/'+i)
    img = cv2.resize(img, (28,28))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = img.flatten()
    x.append(img)
    y.append(0)
    #print(img.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
classifier = Sequential()
## Adding the input layer and frist hidden layer :
classifier.add(Dense(128, activation = 'relu'))
## Adding the Second hidden layer :
classifier.add(Dense(100, activation = 'relu'))
classifier.add(Dense(64, activation = 'relu'))
## Adding the output Layer :
#classifier.add(Dense(64, activation='relu'))
classifier.add(Dropout(0.2))
classifier.add(Dense(1, activation = 'sigmoid'))
```

```python
from keras.optimizers import sgd
```

```python
sgd = SGD(lr = 0.0001, decay = 1e-6, momentum = 0.5, nesterov= True)
```

```python
## Compile the ANN 
classifier.compile(optimizer = sgd, loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
## Fitting the ANN to Training set :
#classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 25)
classifier.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 10)
```

```python
pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred)
```
